```python
import pandas as pd
import numpy as np
```

```python
### load data
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
users = []
for f_name in glob('Students/*.csv'):
    user = pd.read_csv(f_name, parse_dates=['resp_time'], index_col=['resp_time'],date_parser=dateparse)
    users.append(user)

df = pd.concat(users, ignore_index=False)

df.sort_values(by='resp_time').head()
all_data = df.groupby(level=0).mean()
all_data.head(10)
```

```python
all_data.describe()
```

# baseline 1: average stress

```python
stress_day_percnt = len(all_data[all_data['stress_level'] > 2.5]) / len(all_data)
stress_day_percnt
```

baseline 1: average stress level <br>
The average student stress level is 1.305. If we set stress level > 2.5 to be stress, and stress level <2.5 to be non-stress, only 6.25% of all surveyed days that students would generally feel stressed.

# baseline 2: Naive bayes Classifier

```python
all_data['stress_level'].head()
```

```python
# set stress = 1
all_data['stress_or_not'] = np.where(all_data['stress_level'] >= 2.5, 1,0)
all_data['stress_or_not']
all_data.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import sklearn.naive_bayes as sknb

X = all_data.drop(columns = ['stress_level','stress_or_not']).values
y = all_data['stress_or_not']

X_train, X_test, y_train, y_test = train_test_split(X,y)

gnb_model = sknb.GaussianNB()
gnb_model.fit(X_train,y_train)

y_pred = gnb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {0:.2%}".format(accuracy))
```
