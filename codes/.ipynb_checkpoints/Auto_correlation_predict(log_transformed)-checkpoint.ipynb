# TS Analysis - Prediction (log transformed "Stress")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.stattools import acf, pacf 
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
```

# Data Processing

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
users = []
for f_name in glob('Students/*.csv'):
    user = pd.read_csv(f_name, parse_dates=['resp_time'], index_col=['resp_time'],date_parser=dateparse)
    users.append(user)

df = pd.concat(users, ignore_index=False)
df.head()
```

```python
df.sort_values(by='resp_time').head()
all_data = df.groupby(level=0).mean()
all_data.head()
```

# ts prediction

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics

    rolmean = timeseries.rolling(3).mean()  # data量太少,故调小一点sliding window size
    rolstd = timeseries.rolling(3).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.savefig('pics/rolling_means&sd.png')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
ts_stress = all_data[['stress_level']].stress_level
moving_avg = ts_stress.rolling(3).mean()

plt.plot(ts_stress)
plt.plot(moving_avg, color='red')
plt.title('Stress trend with moving average')
plt.savefig('pics/moving_average.png')
```

```python
test_stationarity(ts_stress)
```

```python
# log transform Stress data
ts_log = np.log(ts_stress)  
plt.plot(ts_log)
plt.title('log transformed Stress trend')
```

```python
ts_moving_avg_diff = ts_stress - moving_avg
ts_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_moving_avg_diff)
```

```python
# 指数加权移动平均法(用log value来计算)
expwighted_avg = ts_log.ewm(3).mean()  # set sliding window size to be 3
plt.plot(ts_log)
plt.plot(expwighted_avg, color='red')
plt.title('expweighted_avg (log_transformed)')
```

```python
# 定义指数衰减量 (log transformed)
ts_log_ewma_diff = ts_log - expwighted_avg
test_stationarity(ts_ewma_diff)
```

```python
# 差分 (log transform)
ts_log_diff = ts_log - ts_log.shift()
plt.plot(ts_log_diff)
```

```python
#### 验证差分效果(log transform) 
ts_log_diff = ts_log_diff.replace([np.inf,-np.inf],np.nan).dropna()  # 处理 log transform之后的 missing values跟NA value
test_stationarity(ts_log_diff)
```

```python
#### log transformed
from statsmodels.tsa.stattools import acf, pacf  # 时间序列的自回归函数和部分自回归函数
lag_acf_log = acf(ts_log_diff, nlags=3)
lag_pacf_log = pacf(ts_log_diff, nlags=3, method='ols')
```

```python
#Plot 自相关函数（ACF）：这是时间序列和它自身滞后版本之间的相关性的测试 (log transformed)
plt.subplot(121) 
plt.plot(lag_acf_log)
plt.axhline(y=0,linestyle='--',color='gray')

plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')

plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function (log)')
```

```python
# 处理ts_log的NA跟inf value
ts_log= ts_log.replace([np.inf,-np.inf],np.nan).dropna() 
```

```python
# 法二: plot ACF(自相关)  
# 图中 x 轴表示滞后值，y 轴上 -1 和 1 之间则表现了这些滞后值的正负相关性。
# 蓝色区域中的点表示统计学显着性。滞后值为 0 相关性为 1 的点表示观察值与其本身 100% 正相关。
# 这个分析为后续的比较过程提供了一个很好的基准。

from statsmodels.graphics.tsaplots import plot_acf
# plot_acf(ts_stress)
plot_acf(ts_log)
plt.savefig('pics/ACF(log).png')
plt.show()
```

```python
#Plot PACF:  # 部分自相关函数(PACF):这是时间序列和它自身滞后版本之间的相关性测试，但是是在预测（已经通过比较干预得到解释）的变量后。
plt.subplot(122)
plt.plot(lag_pacf_log)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function (log)')
plt.tight_layout()
```

# Fit model + Plot

```python
### AR(log transform) - 自回归（AR）模型
from statsmodels.tsa.arima_model import ARIMA
# model = ARIMA(ts_log, order=(2, 1, 0)) 
# model = ARIMA(ts_log, order=(1, 1, 0))  # set lag = 1
model = ARIMA(ts_log, order=(3, 1, 0))  

results_AR = model.fit(disp=-1)
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('AR (RSS: %.4f)'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

```python
# 移动平均数（MA）模型 (log transform)

model = ARIMA(ts_log, order=(0, 1, 2))   # set sliding window = 2
results_MA = model.fit(disp=-1)  

plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('MA (RSS: %.4f)'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

```python
# 组合模型 (log transform)

# model = ARIMA(ts_log, order=(2, 1, 2))
model = ARIMA(ts_log, order=(1, 1, 2))  # lag = 1, sliding_window = 2
# model = ARIMA(ts_log, order=(7, 1, 2))  

results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('ARIMA (RSS: %.4f)'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

# 模型导回原始区间

```python
# ARIMA(组合模型) 倒回到原始区间  
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(predictions_ARIMA_diff.head())
```

```python
# MA 倒回到原始区间 
predictions_MA_diff = pd.Series(results_MA.fittedvalues, copy=True)
print(predictions_MA_diff.head())
```

```python
# AR 倒回到原始区间 
predictions_AR_diff = pd.Series(results_AR.fittedvalues, copy=True) 
print (predictions_AR_diff.head())
```

# 模型累计总和

```python
# AR 累计总和
predictions_AR_diff_cumsum = predictions_AR_diff.cumsum()
print (predictions_AR_diff_cumsum.head())
```

```python
# MA 累计总和
predictions_MA_diff_cumsum = predictions_MA_diff.cumsum()
print (predictions_MA_diff_cumsum.head())
```

```python
# ARIMA 累计总和
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print (predictions_ARIMA_diff_cumsum.head())
```

```python
### ARIMA结果（log transform） 
predictions_ARIMA_log = pd.Series(ts_log.ix[0], index=ts_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
### MA结果
predictions_MA = pd.Series(ts_log.ix[0], index=ts_log.index)
predictions_MA = predictions_MA.add(predictions_MA_diff_cumsum,fill_value=0)
predictions_MA.head()
```

```python
### AR结果
predictions_AR = pd.Series(ts_log.ix[0], index=ts_log.index)
predictions_AR = predictions_AR.add(predictions_AR_diff_cumsum,fill_value=0)
predictions_AR.head()
```

# Use model to predict stress

```python
### ARIMA模型 predict stress
predictions_ARIMA_log = np.exp(predictions_ARIMA_log)
plt.plot(ts_log)
plt.plot(predictions_ARIMA_log)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA_log-ts_log)**2)/len(ts_log)))
```

```python
# MA模型 predict stress
predictions_MA = np.exp(predictions_MA)
plt.plot(ts_log)
plt.plot(predictions_MA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_MA-ts_log)**2)/len(ts_log)))
```

```python
# AR模型 predict stress
predictions_ = np.exp(predictions_AR)
plt.plot(ts_log)
plt.plot(predictions_AR)
plt.title('AR_log (RMSE: %.4f)'% np.sqrt(sum((predictions_AR-ts_log)**2)/len(ts_log)))
plt.savefig('pics/AR_log_pred')

# lag = 7 , 0.8068
# lage = 3, 0.8068
# lag = 2, 0.7561
# lag = 1, 0.6704
```
