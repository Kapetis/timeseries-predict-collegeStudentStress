# TS Analysis - Prediction (original "Stress")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.stattools import acf, pacf 
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
```

# Data Processing

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
users = []
for f_name in glob('Students/*.csv'):
    user = pd.read_csv(f_name, parse_dates=['resp_time'], index_col=['resp_time'],date_parser=dateparse)
    users.append(user)

df = pd.concat(users, ignore_index=False)
df.head()
```

```python
df.sort_values(by='resp_time').head()
all_data = df.groupby(level=0).mean()
all_data.head()
```

# ts prediction

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics

    rolmean = timeseries.rolling(3).mean()  # data量太少,故调小一点sliding window size
    rolstd = timeseries.rolling(3).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.savefig('pics/rolling_means&sd.png')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
# ts_stress = all_data[['stress_level']]
ts_stress = all_data[['stress_level']].stress_level
moving_avg = ts_stress.rolling(3).mean()

plt.plot(ts_stress)
plt.plot(moving_avg, color='red')
plt.title('Stress trend with moving average')
plt.savefig('pics/moving_average.png')
```

```python
test_stationarity(ts_stress)
```

```python
ts_moving_avg_diff = ts_stress - moving_avg
ts_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_moving_avg_diff)
```

```python
# 指数加权移动平均法
expwighted_avg = ts_stress.ewm(3).mean()  # set sliding window size to be 3
plt.plot(ts_stress)
plt.plot(expwighted_avg, color='red')
plt.title('expweighted_avg')
```

```python
# 定义指数衰减量
ts_ewma_diff = ts_stress - expwighted_avg
test_stationarity(ts_ewma_diff)
```

```python
# 差分 -> 消除季节性和趋势
ts_diff = ts_stress - ts_stress.shift()
plt.plot(ts_diff)
```

```python
# 验证差分效果(有无更平稳)
ts_diff.dropna(inplace=True)
test_stationarity(ts_diff)    # p-value = 0.027704 说明基本平稳
```

```python
from statsmodels.tsa.stattools import acf, pacf  # 时间序列的自回归函数和部分自回归函数

lag_acf = acf(ts_diff, nlags=3)
lag_pacf = pacf(ts_diff, nlags=3, method='ols')
```

```python
#Plot ACF:  # 自相关函数（ACF）：这是时间序列和它自身滞后版本之间的相关性的测试
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')

plt.axhline(y=-1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')

plt.axhline(y=1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
# 法二: plot ACF(自相关)  
# 图中 x 轴表示滞后值，y 轴上 -1 和 1 之间则表现了这些滞后值的正负相关性。
# 蓝色区域中的点表示统计学显着性。滞后值为 0 相关性为 1 的点表示观察值与其本身 100% 正相关。
# 这个分析为后续的比较过程提供了一个很好的基准。

from statsmodels.graphics.tsaplots import plot_acf
plot_acf(ts_stress)
plt.savefig('pics/ACF.png')
plt.show()
```

```python
#Plot PACF:  # 部分自相关函数(PACF):这是时间序列和它自身滞后版本之间的相关性测试，但是是在预测（已经通过比较干预得到解释）的变量后。
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

# Fit model + Plot

```python
from statsmodels.tsa.arima_model import ARIMA
# 自回归（AR）模型

# model = ARIMA(ts_stress, order=(2, 1, 0)) 
model = ARIMA(ts_stress, order=(7, 1, 0))    # 17.3835 min
# model = ARIMA(ts_stress, order=(1, 1, 0))   

results_AR = model.fit(disp=-1)
plt.plot(ts_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_diff)**2))

### 设lag = 6, sliding window = 0 | 为控制变量跟之后的model有可比性
### vol.2版: 设lag = 1, sliding window = 0  | RSS: 29.68
```

当lag = 6或7的时候，auto correlation正相关, 所以order (p,d,q) 中p可设为6.  <br>
q(sliding window) = 0 --> RSS: 17.8818 <br>
q = 1 --> RSS: 17.6651 <br>
q = 2 --> *error: SVD not converge <br>
q = 3 --> RSS: 11630.03 |很大, 但是基本把原来的stress trend明显突出 <br>

```python
# 移动平均数（MA）模型
# model = ARIMA(ts_stress, order=(0, 1, 1))   
model = ARIMA(ts_stress, order=(0, 1, 2))   # set sliding_Window = 2, RSS min = 22.2106
results_MA = model.fit(disp=-1)  

plt.plot(ts_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_diff)**2))

### 当lag=0,则auto-correlation为1，作为控制变量去比较不同model； 设sliding window = 1 | RSS: 25.1269
### vol.2； 设sliding window = 2 (RSS: 22.2106); 设sliding window = 3 (RSS: 22.2129); 
```

```python
# 组合模型
# model = ARIMA(ts_stress, order=(6, 1, 1))
model = ARIMA(ts_stress, order=(7, 1, 2))
# model = ARIMA(ts_stress, order=(1, 1, 2))

results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_diff)**2))

## 既设lag = 6,又设sliding window = 1 --> 结合起来 （组合模型）
## vol.2: 设lag = 1,又设sliding window = 2  | RSS: 22.37
```

小结：lag = 6, sliding window = 1 <br>
组合模型的表现好于单个model <br>
ARIMA(RSS:17.6651) > AR(RSS:17.8818) > MA(RSS:25.1269) <br>

vol2.小结：lag = 1, sliding window = 2 <br>
组合模型的表现好于单个model <br>
MA(RSS:22.2106) > ARIMA(RSS:22.3787) > AR(RSS:29.6872) <br>

# 模型导回原始区间

```python
# ARIMA(组合模型) 倒回到原始区间 
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(predictions_ARIMA_diff.head())
```

```python
# MA 倒回到原始区间 
predictions_MA_diff = pd.Series(results_MA.fittedvalues, copy=True)
print(predictions_MA_diff.head())
```

```python
# AR 倒回到原始区间 
predictions_AR_diff = pd.Series(results_AR.fittedvalues, copy=True) 
print (predictions_AR_diff.head())
```

# 模型累计总和

```python
# AR 累计总和
predictions_AR_diff_cumsum = predictions_AR_diff.cumsum()
print (predictions_AR_diff_cumsum.head())
```

```python
# MA 累计总和
predictions_MA_diff_cumsum = predictions_MA_diff.cumsum()
print (predictions_MA_diff_cumsum.head())
```

```python
# ARIMA 累计总和
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print (predictions_ARIMA_diff_cumsum.head())
```

```python
### ARIMA结果
predictions_ARIMA = pd.Series(ts_stress.ix[0], index=ts_stress.index)
predictions_ARIMA = predictions_ARIMA.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA.head()
```

```python
### MA结果
predictions_MA = pd.Series(ts_stress.ix[0], index=ts_stress.index)
predictions_MA = predictions_MA.add(predictions_MA_diff_cumsum,fill_value=0)
predictions_MA.head()
```

```python
### AR结果
predictions_AR = pd.Series(ts_stress.ix[0], index=ts_stress.index)
predictions_AR = predictions_AR.add(predictions_AR_diff_cumsum,fill_value=0)
predictions_AR.head()
```

# Use model to predict stress

```python
# ARIMA模型 predict stress
predictions_ARIMA = np.exp(predictions_ARIMA)
plt.plot(ts_stress)
plt.plot(predictions_ARIMA)
plt.title('ARIMA (RMSE: %.4f)'% np.sqrt(sum((predictions_ARIMA-ts_stress)**2)/len(ts_stress)))

### 预测值跟实际值差太多
```

```python
# MA模型 predict stress
predictions_MA = np.exp(predictions_MA)
plt.plot(ts_stress)
plt.plot(predictions_MA)
plt.title('MA (RMSE: %.4f)'% np.sqrt(sum((predictions_MA-ts_stress)**2)/len(ts_stress)))
```

```python
# AR模型 predict stress
predictions_ = np.exp(predictions_AR)
plt.plot(ts_stress)
plt.plot(predictions_AR)
plt.title('AR (RMSE: %.4f)'% np.sqrt(sum((predictions_AR-ts_stress)**2)/len(ts_stress)))
```
