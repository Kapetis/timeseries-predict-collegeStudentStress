```python
def get_user_index(folder_path):
    file_paths = []
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        file_paths.append(file_path)
        
    users = []
    for path in file_paths:
#         user_id = path.split('.')[0][-2:]
        user_id = path.split('.')[0][-3:]
        users.append(user_id)
        
    return users
```

```python
def get_feature_df(users,feature_data, col_name):
    data = {}
    data['user_id'] = users

    data[col_name] = feature_data
    
    df = pd.DataFrame(data).sort_values('user_id')
    return df
```

```python
# read in user "stress" data one by one
import os

stress_level_lst = []
for filename in os.listdir('Data/Stress'):
    with open(os.path.join('Data/Stress',filename)) as f:
        content = f.read()
        df = pd.read_json(content)
        if 'level' in df.columns:
            mean_stress_level = np.mean(df.level)
            stress_level_lst.append(mean_stress_level)
        else:
            mean_stress_level = np.nan
            stress_level_lst.append(mean_stress_level)
        
        
len(stress_level_lst)
```

```python
stress_users = get_user_index('Data/Stress')
stress_df = get_feature_df(stress_users,stress_level_lst,'stress')
stress_df.head()
```

```python
# Sleep 
sleep_hour_lst = []
for filename in os.listdir('Data/Sleep'):
    with open(os.path.join('Data/Sleep', filename)) as f:
        content = f.read()
        df = pd.read_json(content)
        if 'hour' in df.columns:
            avg_sleep_hour = np.mean(df.hour)
            sleep_hour_lst.append(avg_sleep_hour)
        else:
            avg_sleep_hour = np.nan
            sleep_hour_lst.append(avg_sleep_hour)

len(sleep_hour_lst)  # 49
```

```python
sleep_users = get_user_index('Data/Sleep')
sleep_df = get_feature_df(sleep_users,sleep_hour_lst,'sleep')
sleep_df.head()
```

```python
# activity

def get_activity_data(field):
    value_lst = []
    
    for filename in os.listdir('Data/Activity'):
        with open(os.path.join('Data/Activity', filename)) as f:
            content = f.read()
            df = pd.read_json(content)
            df = df.replace('null', np.NaN) # data cleaing

            if field in df.columns:
                avg_value = np.mean(df[field])
                value_lst.append(avg_value)
                
            else:
                avg_value = np.nan
                value_lst.append(avg_value)
    
    return value_lst

activities = {}
working_alone = get_activity_data('working')
working_others = get_activity_data('other_working')
relaxing_alone = get_activity_data('relaxing')
relaxing_others = get_activity_data('other_relaxing')
```

```python
#### get activity_user
activity_users = get_user_index('Data/Activity')

#### get activity_df
# activity_df = get_feature_df(activity_users,social_number_lst,'social') 

data = {}
data['user_id'] = activity_users
data['activity__working_alone'] = working_alone
data['activity__working_others'] = working_others
data['activity__relaxing_alone'] = relaxing_alone
data['activity__relaxing_others'] = relaxing_others

activity_df = pd.DataFrame(data).sort_values('user_id')
activity_df.head()
```

```python
# behavior
def get_behavior_data(field):
    value_lst = []
    for filename in os.listdir('Data/Behavior'):
        with open(os.path.join('Data/Behavior', filename)) as f:
            content = f.read()
            df = pd.read_json(content)
            df = df.replace('null', np.NaN)
            
            if field in df.columns:
                avg_value = np.mean(df[field].astype('float64'))
                value_lst.append(avg_value)
                
            else:
                avg_value = np.nan
                value_lst.append(avg_value)
                
    return value_lst
```

```python
#### get behavior_user
behavior_users = get_user_index('Data/Behavior')

#### get activity_df
data = {}
data['user_id'] = activity_users
data['behavior__anxious'] = get_behavior_data('anxious')
data['behavior__calm'] = get_behavior_data('calm')
data['behavior__conventional'] = get_behavior_data('conventional')
data['behavior__critical'] = get_behavior_data('critical')
data['behavior__dependable'] = get_behavior_data('dependable')
data['behavior__disorganized'] = get_behavior_data('disorganized')
data['behavior__enthusiastic'] = get_behavior_data('enthusiastic')
data['behavior__experiences'] = get_behavior_data('experiences')
data['behavior__reserved'] = get_behavior_data('reserved')
data['behavior__sympathetic'] = get_behavior_data('sympathetic')

behavior_df = pd.DataFrame(data).sort_values('user_id')
behavior_df.head()
```

```python
# mood

def get_mood_data(field):
    value_lst = []
    for filename in os.listdir('Data/Mood'):
        with open(os.path.join('Data/Mood', filename)) as f:
            content = f.read()
            df = pd.read_json(content)
            df = df.replace('null', np.NaN)
            
            if field in df.columns:
                avg_value = np.mean(df[field].astype('float64'))
                value_lst.append(avg_value)
                
            else:
                avg_value = np.nan
                value_lst.append(avg_value)
                
    return value_lst
```

```python
# get mood users
mood_users = get_user_index('Data/Mood')

# build mood_df
data = {}
data['user_id'] = mood_users
data['mood__happy'] = get_mood_data('happy')
data['mood__sad'] = get_mood_data('sad')

mood_df = pd.DataFrame(data).sort_values('user_id')
mood_df.head()
```
